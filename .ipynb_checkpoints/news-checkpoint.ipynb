```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import lxml
import re
```

```python
results = requests.get('https://en.as.com/en/2020/05/30/other_sports/1590815660_838720.html')
results
```

```python
soup = BeautifulSoup(results.content, 'html.parser')
```

```python
listings = soup.find_all('div', {'class':'ev-unit'})
```

```python
listings
```

```python
def if_exists(x):
    if x is not None:
        return x.text
    else:
        return 'empty'
```

```python
news = []
for new in listings:

    title = if_exists(new.find('h3').replace('\xa0', ' '))
    newss = new.find('p').text.replace('\xa0', '')
    news.append([title, newss])
news  
```
