```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import lxml
import re
```

```python
results = requests.get('https://meqasa.com/houses-for-rent-in-ghana')
results
```

```python
soup = BeautifulSoup(results.content, 'lxml')
```

```python
listings = soup.find_all('div', {'class':'mqs-featured-prop-inner-wrap'})
```

```python
def if_exists(x):
	if type(x) == str:
		if x is not None:
			return x.text
		else:
			return 'None'
	else:
		x = 0
		return x
```

```python
real_estate = []
for tag in listings:
    property_name = if_exists(tag.find('h2')).replace('\n', '')
    bed =  tag.find('li', {'class':'bed'}).text.replace('\n', '')
    beds = int(bed)
    garages = if_exists(tag.find('li', {'class':'garage'}))
    showers = tag.find('li', {'class':'shower'}).text.replace('\n', '')
    area = if_exists(tag.find('li', {'class':'area'}))
    description = if_exists(tag.find_all('p')[1])
    price = if_exists(tag.find('p', {'class':'h3'})).replace('\nPrice', '').replace('/ month\n', '').replace('[Price disclosed on request] \n', '')
    currency = tag.find('p', {'class':'h3'}).text.replace('\nPrice', '').replace('/ month\n', '').replace('[Price disclosed on request] \n', '')
    currency = re.sub('\d', '', currency)
    rent = tag.find('p', {'class':'h3'}).text.replace('\nPrice$', '').replace('\nPriceGH₵', '').replace('\n', '').replace('PricePricedisclosedonrequest', '')
    c = re.sub('\d.....\W', '', rent)
    d = re.sub('\d....\W', '', c)
    rentperiod = re.sub('\W', '', d)
    url = tag.find('a')['href']
    url = 'https://meqasa.com/houses-for-rent-in-ghana' + url
    address = tag.find('h2').text.replace('\n', '').split('at')[1]
    timeposted = if_exists(tag.find('p', {'class': 'wsnr'}))
    

    
        
    real_estate.append(
        {
            'property': property_name,             
            'showers': showers,
            'beds': beds,            
            'garages': garages,
            'area': area,
            'description': description,
            'price': price,
            'currency': currency,
            'rent period': rentperiod,
            'url': url,  
            'address': address,
            'time posted': timeposted,
            
        }
    )
```

```python
df = pd.DataFrame(real_estate)
df
```
